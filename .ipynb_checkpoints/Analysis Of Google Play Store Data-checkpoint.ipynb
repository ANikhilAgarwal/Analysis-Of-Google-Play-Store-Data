```python
# Here We are checking the Percentage of NAN Values present in each column.
# 1. Make the list of features having missing values
features_with_nan = [
    features for features in datasetnomiss.columns
    if datasetnomiss[features].isnull().sum() > 1
]


def list_nan():
    # print the feature name and the persentage of missing values
    for feature in features_with_nan:
        print(color.BOLD + feature,
              ": ",
              color.END + color.RED,
              np.round(datasetnomiss[feature].isnull().mean() * 100, 4),
              "% ",
              color.END,
              "Missing Values",
              sep='')


list_nan()
```

```python
def missing_table(features_with_nan):
    missing_count = (
        datasetnomiss[features_with_nan].isnull().sum()).sort_values(
            ascending=False)
    missing_percentage = missing_count / len(
        datasetnomiss[features_with_nan]) * 100
    missing_data = pd.DataFrame({
        "Count of missing values":
        missing_count,
        "Percentage of missing values":
        missing_percentage
    })
    return missing_data


missing_table(features_with_nan)
```

# Encoding and Formatting

```python
datasetnomiss[datasetnomiss['Content Rating'] == "Adults only 18+"]
```

```python
# 1) Change the RatingCount dtype to int
datasetencode['Rating Count'] = dataset['Rating Count'].astype(int)

# 2) Round up the Rating, Price, and Average Installs columns to 2 decimal points
datasetencode[['Rating', 'Price', 'Average Installs'
               ]] = dataset[['Rating', 'Price', 'Average Installs']].round(2)

# 3) Reformat release and update dates to show months and year (i.e. Release Month and Release Year, etc.)
datasetencode['Released'] = pd.to_datetime(dataset['Released'],
                                           format='%b %d, %Y')
datasetencode['Last Updated'] = pd.to_datetime(dataset['Last Updated'],
                                               format='%b %d, %Y')
```

```python
# 4) Change the 'Size' data in KB to MB and change the type 'object' to 'float'

# found value with '1,000+' in one of record, remove it from data_frame as uncertain whether it is 'M' or 'k'
sizes = [i for i in datasetencode['Size']]


def clean_sizes(sizes_list):
    """
    As sizes are represented in 'M' and 'k', we remove 'M'
    and convert 'k'/kilobytes into megabytes
    """
    cleaned_data = []
    for size in sizes_list:
        if 'M' in size:
            size = size.replace('M', '')
            size = size.replace(',', '')
            size = float(size)
        elif 'k' in size:
            size = size.replace('k', '')
            size = size.replace(',', '')
            size = float(size)
            size = size / 1024  # 1 megabyte = 1024 kilobytes
        elif 'G' in size:
            size = size.replace('G', '')
            size = size.replace(',', '')
            size = float(size)
            size = size * 1024  # 1 Gigabyte = 1024 Megabyte
        # representing 'Varies with device' with value 0
        elif 'Varies with device' in size:
            size = float(0)
        cleaned_data.append(size)
    return cleaned_data


datasetencode['Size'] = clean_sizes(sizes)
datasetencode['Size'] = datasetencode['Size'].astype(float)
```

### Visualizing Number of applications released with respect to different years for "Actual dataset", "Non-gaming dataset" and "Gamaing dataset 

```python
categorydataset_column = [
    "Actual dataset", "Non-gaming dataset", "Gamaing dataset"
]
dd = widgets.Dropdown(options=categorydataset_column,
                      value=categorydataset_column[0],
                      description="Select Category: ",
                      style={'description_width': 'initial'})


def draw_countplot(column):
    if column == "Actual dataset":
        #get the release year
        years = dataset['Released'].apply(lambda x: x.year)
    elif column == "Non-gaming dataset":
        #get the release year
        years = appdf['Released'].apply(lambda x: x.year)
    elif column == "Gamaing dataset":
        #get the release year
        years = gamesdf['Released'].apply(lambda x: x.year)
    release_years = years.value_counts()
    release_years = release_years.reset_index()
    release_years.rename(columns={"index": "year"}, inplace=True)
    release_years.sort_values(by='year', ascending=False, inplace=True)

    #plot the graph
    plt.plot(release_years['year'], release_years['Released'], marker='o')
    plt.xlabel('Year')
    plt.ylabel('Number of Apps Released')
    plt.xticks([i for i in range(2010, 2022)], rotation=30)
    plt.xlim([2010, 2020])
    plt.title('Number of Apps Released per Year')
    plt.show()


interact(draw_countplot, column=dd)
```

### Boxplot for rating for each category

```python
categorydataset_column = [
    "Actual dataset", "Non-gaming dataset", "Gamaing dataset"
]
dd = widgets.Dropdown(options=categorydataset_column,
                      value=categorydataset_column[0],
                      description="Select Category: ",
                      style={'description_width': 'initial'})


def draw_countplot(column):
    if column == "Actual dataset":
        dt = dataset
    elif column == "Non-gaming dataset":
        dt = appdf
    elif column == "Gamaing dataset":
        dt = gamesdf
    plt.figure(figsize=(12, 12))

    median_ratings = dt[dt['Rating'] > 0].groupby(
        'Category')['Rating'].median()[:]
    plot_order = median_ratings.sort_values(ascending=False).index
    plt.title('Ratings per Category for %s' % column)
    sns.boxplot(x='Rating',
                y='Category',
                data=dt[dt['Rating'] > 0],
                order=plot_order)

    plt.show()


interact(draw_countplot, column=dd)
```

    the conclusion for all three(Actual dataset, Non-gaming dataset, Gaming dataset) can be made with the number of installs on the play store
    Actual Dataset & Non-Gaming Dataset:
    One thing to point out is the fact that the Events category holds the highest mean rating, while having the least number of installs on the Google Play Store (check the graph directly above). Actually, if you were to think about it, that would make perfect sense, since there are less number of apps that result in a regression of the rating to the mean, causing slightly 'inflated' ratings for categories with lesser apps (although this does not apply to most categories).
    
    Gaming Dataset:
    The above statement also applies for the gaming category of applications where Word category holds the highest mean rating, while being in the bottom for the number of installs on the Google Play Store
